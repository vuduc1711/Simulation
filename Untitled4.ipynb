```python
def group_time(data,time):    
    data_max = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dT'%time))['Price'].max()).reset_index()
    data_min = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dT'%time))['Price'].min()).reset_index()
    data_open = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dT'%time))['Price'].first()).reset_index()
    data_close = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dT'%time))['Price'].last()).reset_index()

    data_max.rename(columns={'Price':'High'},inplace=True)
    data_min.rename(columns={'Price':'Low'},inplace=True)
    data_open.rename(columns={'Price':'Open'},inplace=True)
    data_close.rename(columns={'Price':'Close'},inplace=True)

    data_full = pd.merge(data_max,data_min,how='inner',left_on='Date',right_on='Date')
    data_full = pd.merge(data_full,data_open,how='inner',left_on='Date',right_on='Date')
    data_full = pd.merge(data_full,data_close,how='inner',left_on='Date',right_on='Date')
    
    return data_full
```

```python
def group_time(data,time):    
    data_max = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dD'%time))['Price'].max()).reset_index()
    data_min = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dD'%time))['Price'].min()).reset_index()
    data_open = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dD'%time))['Price'].first()).reset_index()
    data_close = pd.DataFrame(data.groupby(pd.Grouper(key = 'Date', freq='%dD'%time))['Price'].last()).reset_index()

    data_max.rename(columns={'Price':'High'},inplace=True)
    data_min.rename(columns={'Price':'Low'},inplace=True)
    data_open.rename(columns={'Price':'Open'},inplace=True)
    data_close.rename(columns={'Price':'Close'},inplace=True)

    data_full = pd.merge(data_max,data_min,how='inner',left_on='Date',right_on='Date')
    data_full = pd.merge(data_full,data_open,how='inner',left_on='Date',right_on='Date')
    data_full = pd.merge(data_full,data_close,how='inner',left_on='Date',right_on='Date')
    
    return data_full
```

```python
def chart(data,stock):
    layout = dict(
    title = stock,
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=5,
                     label='15M',
                     step='minute',
                     stepmode='backward'),
                dict(count=10,
                     label='30m',
                     step='minute',
                     stepmode='backward'),
                dict(count=1,
                    label='1h',
                    step='hour',
                    stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type="date", categoryorder='category ascending',
        rangebreaks=[
            dict(bounds=["sat", "mon"]), 
            dict(bounds=[14.5, 9.25], pattern="hour"),  # hide hours outside of 9.30am-4pm
    #             dict(values=["2021-01-21"])  # hide holidays (Christmas and New Year's, etc)
            ]       
        ),yaxis = dict(
         fixedrange = False),autosize=False,
        width=1000,
        height=1000)

    
    fig = go.Figure(data=[go.Candlestick(x=data['Date'],
                    open=data['Open'],
                    high=data['High'],
                    low=data['Low'],
                    close=data['Close'])],layout=layout)
    
    for parameter in ['moving_average', 'bol_lower', 'bol_upper']:
        fig.add_trace(go.Scatter(
            x = data.Date,
            y = data[parameter],
            showlegend = False,
            line_color = 'gray',
            mode='lines',
            line={'dash': 'dash'},
            marker_line_width=2, 
            marker_size=10,
            opacity = 0.8))
        

    fig.show()
#     return fig
```
